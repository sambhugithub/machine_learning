```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
```

```python
data = pd.read_csv("D:\\exe_file\\machine_learning\\csv_files\\Social_Network_Ads.csv")
data
```

```python
real_x=data.iloc[:,[2,3]].values
real_y=data.iloc[:,4].values
real_x
```

```python
training_x,test_x,training_y,test_y = train_test_split(real_x,real_y,test_size=0.25,random_state=0)
training_x
```

```python
scaler=StandardScaler()
training_x = scaler.fit_transform(training_x)
test_x = scaler.fit_transform(test_x)
training_x
```

```python
cls=SVC(kernel='linear',random_state=0)
cls.fit(training_x,training_y)
```

```python
y_pred=cls.predict(test_x)
y_pred
```

```python
cm=confusion_matrix(test_y,y_pred)
cm
```

```python
x_set,y_set=training_x,training_y
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,cls.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0],x_set[y_set == j,1],
                c=ListedColormap(('red','green'))(i),label=j)
plt.title('svm for training data')
plt.xlabel('age')
plt.ylabel('estimated salary')
plt.legend()
plt.show()
```
